```python
import math
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)

import matplotlib.pyplot as plt
```

```python
dfKKRvRCB = pd.read_csv('../../Material/KKRvRCB_2008.csv')
```

```python
def getPartners(dfp, batter1, batter2):
    
    partners = [batter1, batter2]
    partners.sort()
    
    return (partners[0] + '_' + partners[1])
```

```python
# df['runs_off_bat'].count(), df[df['extras' != 0]].count()
```

```python
# Create a new column of total_runs & balls
df['total_runs'] = df['runs_off_bat'] + df['extras']
# df['total_balls'] = df['runs_off_bat'] - df['extras']

# Create a new column of striker_non Striker
df['partners'] = df.apply(lambda x: getPartners(x, x['striker'], x['non_striker']), axis = 1)
df.head()
```
