### KKR vs RCB Batting Partnership Analysis
- Display the batting partnerships in different possible ways
- Partership analysis
    - Get player stats (individual and with partner)
    - Get Partnership Total Runs, Strike Rate, Contribution

```python
import math
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)

import matplotlib.pyplot as plt

import seaborn as sns
import itertools
```

```python
dfKKRvRCB = pd.read_csv('../../Material/KKRvsRCB_2008_PS.csv')
```

```python
df = dfKKRvRCB.copy()
```

```python
df['partners'] = df['batter1'] + ' & ' + df['batter2']
df
```

```python
dfKKR = df[df.team == 'Kolkata Knight Riders']
dfRCB = df[df.team == 'Royal Challengers Bangalore']

dfKKR = dfKKR.iloc[::-1]
dfRCB = dfRCB.iloc[::-1]
```

```python
dfKKR.reset_index(inplace = True, drop = True)
dfKKR.head()
```

```python
dfRCB.reset_index(inplace = True, drop = True)
dfRCB.head()
```

```python
def plotBarGraph(dfCur):
    
    # set width of bar
    barWidth = 0.25
    fig, ax = plt.subplots(figsize =(12, 8))

    plt.title(dfCur.team[0] + " Partnership Analysis")

    # set height of bar
    batter1runs = dfCur['batter1_runs'].values
    batter2runs = dfCur['batter2_runs'].values
    extras = dfCur['extras'].values
    totalruns = dfCur['total_runs'].values

    # Set position of bar on X axis
    br1 = np.arange(len(batter1runs))
    br2 = [x + barWidth for x in br1]
    br3 = [x + barWidth for x in br2]

    # Make the plot
    plt.bar(br1, batter1runs, color ='r', width = barWidth,
            edgecolor ='grey', label ='batter1 runs')
    plt.bar(br2, batter2runs, color ='g', width = barWidth,
            edgecolor ='grey', label ='batter2 runs')
    plt.bar(br3, extras, color ='b', width = barWidth,
            edgecolor ='grey', label ='extras')
 

    # Adding Xticks
    plt.xlabel('Partnership runs Breakup', fontweight ='bold', fontsize = 15)
    plt.ylabel('Runs', fontweight ='bold', fontsize = 15)
    plt.xticks([r + barWidth for r in range(len(batter1runs))],
            dfCur['partners'].values, rotation=90)

    plt.legend()
    plt.show()
```

```python
plotBarGraph(dfKKR)
```

```python
plotBarGraph(dfRCB)
```

```python
def addlabels(x,y,z,adj):
    for i in range(len(x)):
        plt.text(y[i]+adj, i-.05, str(y[i])+' ('+str(z[i])+')',
                 bbox = dict(facecolor = 'red', alpha =.5))

def plotStackedBarGraph(dfCur):
    # Stacked bar chart
    barHeight = 0.25
    batter1runs = dfCur['batter1_runs'].values

    ax = dfCur[['batter1_runs', 'batter2_runs', 'extras']].plot(kind='barh', stacked=True)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='lower left')
    plt.xlabel('Runs in Partnerships')
    plt.title(dfCur.team[0] + ' Partnerships Analysis')

    plt.yticks([r-.25 + barHeight for r in range(len(batter1runs))],
                dfCur['partners'].values)
    
    if dfCur.team[0] == 'Kolkata Knight Riders':
        adj = 2
    else:
        adj = 1
    
    addlabels(dfCur['partners'].values, dfCur['total_runs'].values, dfCur['total_balls'].values, adj)
    
    for s in ['top', 'bottom', 'left', 'right']:
        ax.spines[s].set_visible(False)
    
    for c in ax.containers:

        # Optional: if the segment is small or 0, customize the labels
        labels = [str(int(v.get_width())) if v.get_width() > 0 else '' for v in c]
        
        # remove the labels parameter if it's not needed for customized labels
        ax.bar_label(c, labels=labels, label_type='center')
    
    plt.show()
```

```python
plotStackedBarGraph(dfKKR)
```

```python
plotStackedBarGraph(dfRCB)
```

```python
# Creating dataset

 
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(dfKKR['total_runs'].values, labels = dfKKR['partners'].values)
 
# show plot
plt.show()
```

```python
# Creating dataset
partners = dfKKR['partners'].values
 
runs = dfKKR['total_runs'].values
balls = dfKKR['total_balls'].values

combdt=[]
for i in range(len(partners)):
    combdt.append(partners[i] + ': ' +str(runs[i]) + ' ('+str(balls[i])+')')
                                                           
# Creating explode data
explode = (0.2, 0.1, 0.2, 0.1)
 
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute+1)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(runs,
                                  autopct = lambda pct: func(pct, runs),
                                  explode = explode,
                                  labels = partners,
                                  shadow = True,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="white"))
 
# Adding legend
ax.legend(wedges, combdt,
          title ="KKR Partnerships",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("KKR Partnerships Analysis")
 
# show plot
plt.show()
```
