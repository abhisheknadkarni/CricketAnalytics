## Calculate Player Statistics
- SR = Strike Rate = Scoring rate for each ball
- RPI = Runs Per Innings = Runs scored per each innings including outs/not-outs
- BPB = Balls Per Boundary = No of balls for each six or a four
- BPD = Balls Per Dismissal = Average number of balls faced per each dismissal

```python
# function to get the balls per dismissals, parameters:balls played, no of dismissals
def ballsPerDismissal(balls, dismissals):
    if dismissals > 0:
        return balls/dismissals
    else:
        return balls/1

# function to get the balls per boundaries, parameters: balls played, boundaries scored
def ballsPerBoundary(balls, boundaries):
    if boundaries > 0:
        return balls/boundaries
    else:
        return balls/1
    
# function to get the average of batters
def average(runs, dismissals):
    if dismissals > 1:
        return runs/dismissals
    else:
        return runs

# function to get the strike rotation percentage (1s, 2s & 3s)
def strikeRotation(runs, balls):
    return (100*runs/balls)
    
# create a dataFrame with all the below required player statistics
# dot balls played, singles, twos, threes, fours, sixes scored - done
# runs scored, innings played, balls played, no of dismissals
# strike rate, runs per innings, balls per dismissals, balls per boundaries

# function that returns a new dataFrame with reqd data from the original dataFrame as parameter
def getReqdPlayerStatistics(df, current_phase=''):
    
    if current_phase != '':
        df = df[df.phase == current_phase]
        df.reset_index(inplace = True, drop = True)    
    
    df['isDot'] = df['batsman_runs'].apply(lambda x: 1 if x == 0 else 0)
    df['isOne'] = df['batsman_runs'].apply(lambda x: 1 if x == 1 else 0)
    df['isTwo'] = df['batsman_runs'].apply(lambda x: 1 if x == 2 else 0)
    df['isThree'] = df['batsman_runs'].apply(lambda x: 1 if x == 3 else 0)
    df['isFour'] = df['batsman_runs'].apply(lambda x: 1 if x == 4 else 0)
    df['isSix'] = df['batsman_runs'].apply(lambda x: 1 if x == 6 else 0)
    
    runs = pd.DataFrame(df.groupby(['batsman'])['batsman_runs'].sum().reset_index()).groupby(['batsman'])['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs':'runs'})
    innings = pd.DataFrame(df.groupby(['batsman'])['matchId'].apply(lambda x: len(list(np.unique(x)))).reset_index()).rename(columns = {'matchId':'innings'})
    balls = pd.DataFrame(df.groupby(['batsman'])['matchId'].count()).reset_index().rename(columns = {'matchId':'balls'})
    dismissals = pd.DataFrame(df.groupby(['batsman'])['player_dismissed'].count()).reset_index().rename(columns = {'player_dismissed':'dismissals'})
    
    dots = pd.DataFrame(df.groupby(['batsman'])['isDot'].sum()).reset_index().rename(columns = {'isDot':'dots'})
    ones = pd.DataFrame(df.groupby(['batsman'])['isOne'].sum()).reset_index().rename(columns = {'isOne':'ones'})
    twos = pd.DataFrame(df.groupby(['batsman'])['isTwo'].sum()).reset_index().rename(columns = {'isTwo':'twos'})
    threes = pd.DataFrame(df.groupby(['batsman'])['isThree'].sum()).reset_index().rename(columns = {'isThree':'threes'})
    fours = pd.DataFrame(df.groupby(['batsman'])['isFour'].sum()).reset_index().rename(columns = {'isFour':'fours'})
    sixes = pd.DataFrame(df.groupby(['batsman'])['isSix'].sum()).reset_index().rename(columns = {'isSix':'sixes'})
    
    df = pd.merge(innings, runs, on = 'batsman').merge(balls, on = 'batsman').merge(dismissals, on = 'batsman').merge(dots, on = 'batsman').merge(ones, on = 'batsman').merge(twos, on = 'batsman').merge(threes, on = 'batsman').merge(fours, on = 'batsman').merge(sixes, on = 'batsman')
    
    #Strike Rate
    df['SR'] = df.apply(lambda x: 100*(x['runs']/x['balls']), axis = 1)

    #Runs per innings
    df['RPI'] = df.apply(lambda x: x['runs']/x['innings'], axis = 1)

    #Balls per dismissals
    df['BPD'] = df.apply(lambda x: ballsPerDismissal(x['balls'], x['dismissals']), axis = 1)

    #Balls per boundary
    df['BPB'] = df.apply(lambda x: ballsPerBoundary(x['balls'], (x['fours'] + x['sixes'])), axis = 1)
    
    #Average
    df['AVG'] = df.apply(lambda x: average(x['runs'], x['dismissals']), axis = 1)
    
    #Strike rotation percentage
    df['SRPER'] = df.apply(lambda x: strikeRotation(x['ones']+x['twos']+x['threes'], x['balls']), axis = 1)
    
    #Dot percentage
    df['DOTPER'] = df.apply(lambda x: 100*x['dots']/x['balls'], axis = 1)
    
    return df 
```

```python
# Get Basic Player statistics
df = getReqdPlayerStatistics(deliveries)
df.head()
```

### Performance in different phases of play

```python
def phase(over):
    if over <= 5:
        return 'Powerplay'
    elif over <= 14:
        return 'Middle'
    else:
        return 'Death'
```

```python
wt_sr, wt_rpi, wt_bpd, wt_bpb = 0.38, 0.25, 0.12, 0.26
```

```python
# filter by min 20 innings
powerPlayDF = powerPlayDF[powerPlayDF.innings >= 20]
powerPlayDF.reset_index(inplace = True, drop = True)
```

### Calculation - Normalizing all values to same dimensions

```python
#step1: square of all values
powerPlayDF['calc_SR'] = powerPlayDF['SR'].apply(lambda x: x*x) 
powerPlayDF['calc_RPI'] = powerPlayDF['RPI'].apply(lambda x: x*x) 
powerPlayDF['calc_BPD'] = powerPlayDF['BPD'].apply(lambda x: x*x) 
powerPlayDF['calc_BPB'] = powerPlayDF['BPB'].apply(lambda x: x*x)

sq_sr, sq_rpi, sq_bpd, sq_bpb = np.sqrt(powerPlayDF[['calc_SR','calc_RPI', 'calc_BPD', 'calc_BPB']].sum(axis = 0))

powerPlayDF['calc_SR'] = powerPlayDF['calc_SR'].apply(lambda x: x/sq_sr) 
powerPlayDF['calc_RPI'] = powerPlayDF['calc_RPI'].apply(lambda x: x/sq_rpi) 
powerPlayDF['calc_BPD'] = powerPlayDF['calc_BPD'].apply(lambda x: x/sq_bpd) 
powerPlayDF['calc_BPB'] = powerPlayDF['calc_BPB'].apply(lambda x: x/sq_bpb)

powerPlayDF['calc_SR'] = powerPlayDF['calc_SR'].apply(lambda x: x*wt_sr) 
powerPlayDF['calc_RPI'] = powerPlayDF['calc_RPI'].apply(lambda x: x*wt_rpi) 
powerPlayDF['calc_BPD'] = powerPlayDF['calc_BPD'].apply(lambda x: x*wt_bpd) 
powerPlayDF['calc_BPB'] = powerPlayDF['calc_BPB'].apply(lambda x: x*wt_bpb)

best_sr, worst_sr = max(powerPlayDF['calc_SR']), min(powerPlayDF['calc_SR'])
best_rpi, worst_rpi = max(powerPlayDF['calc_RPI']), min(powerPlayDF['calc_RPI'])
best_bpd, worst_bpd = max(powerPlayDF['calc_BPD']), min(powerPlayDF['calc_BPD'])
best_bpb, worst_bpb = min(powerPlayDF['calc_BPB']), max(powerPlayDF['calc_BPB'])
```

###  Calculation - Comparision against the best & worst values

```python
powerPlayDF['dev_best_SR'] = powerPlayDF['calc_SR'].apply(lambda x: (x-best_sr)*(x-best_sr)) 
powerPlayDF['dev_best_RPI'] = powerPlayDF['calc_RPI'].apply(lambda x: (x-best_rpi)*(x-best_rpi)) 
powerPlayDF['dev_best_BPD'] = powerPlayDF['calc_BPD'].apply(lambda x: (x-best_bpd)*(x-best_bpd)) 
powerPlayDF['dev_best_BPB'] = powerPlayDF['calc_BPB'].apply(lambda x: (x-best_bpb)*(x-best_bpb))

powerPlayDF['dev_best_sqrt'] = powerPlayDF.apply(lambda x: x['dev_best_SR'] + x['dev_best_RPI'] + x['dev_best_BPD'] + x['dev_best_BPB'], axis = 1) 

powerPlayDF['dev_worst_SR'] = powerPlayDF['calc_SR'].apply(lambda x: (x-worst_sr)*(x-worst_sr)) 
powerPlayDF['dev_worst_RPI'] = powerPlayDF['calc_RPI'].apply(lambda x: (x-worst_rpi)*(x-worst_rpi)) 
powerPlayDF['dev_worst_BPD'] = powerPlayDF['calc_BPD'].apply(lambda x: (x-worst_bpd)*(x-worst_bpd)) 
powerPlayDF['dev_worst_BPB'] = powerPlayDF['calc_BPB'].apply(lambda x: (x-worst_bpb)*(x-worst_bpb))

powerPlayDF['dev_worst_sqrt'] = powerPlayDF.apply(lambda x: x['dev_worst_SR'] + x['dev_worst_RPI'] + x['dev_worst_BPD'] + x['dev_worst_BPB'], axis = 1) 
```

### Final Step - Overall score for each player ( values from 0 - 1)

```python
powerPlayDF['score'] = powerPlayDF.apply(lambda x: x['dev_worst_sqrt']/(x['dev_worst_sqrt'] + x['dev_best_sqrt']), axis = 1)
```
