```python
# Create a function to analyse various aspects of every batsmen
```

```python
def setBatsmenDF(strBowlingType=''):
    
    # Set as global variables     
    global dfOrange1
    global dfOrange2
    global dfBlue1
    global dfBlue2
    global dfBlue3
    global dfMaroon
    global dfRed
    
    global dfPercentages
    global batterColors
    
    dfPercentages = pd.DataFrame(columns = ['Batter', 'Balls', 'Middled', 'Missed', 'Edged', 'Left', 'Aggressive', 
                                            'Attacking', 'Defensive', 'Timed', 'Head_Position', 'Footwork', 'Batter_Color'],
        index = ['Orange1', 'Orange2', 'Blue1', 'Blue2', 'Blue3', 'Maroon', 'Red'])
    
    dfOrange1 = dfMain[dfMain['Batsman_Name'] == 'Orange1'].reset_index()
    dfOrange2 = dfMain[dfMain['Batsman_Name'] == 'Orange2'].reset_index()
    dfBlue1 = dfMain[dfMain['Batsman_Name'] == 'Blue1'].reset_index()
    dfBlue2 = dfMain[dfMain['Batsman_Name'] == 'Blue2'].reset_index()
    dfBlue3 = dfMain[dfMain['Batsman_Name'] == 'Blue3'].reset_index()
    dfMaroon = dfMain[dfMain['Batsman_Name'] == 'Maroon'].reset_index()
    dfRed = dfMain[dfMain['Batsman_Name'] == 'Red'].reset_index()
    
    if(strBowlingType != ''):
        # Filter out dataframes for Pace/Spin
        dfOrange1 = dfOrange1[dfOrange1.Bowling_Type == strBowlingType].reset_index()
        dfOrange2 = dfOrange2[dfOrange2.Bowling_Type == strBowlingType].reset_index()
        dfBlue1 = dfBlue1[dfBlue1.Bowling_Type == strBowlingType].reset_index()
        dfBlue2 = dfBlue2[dfBlue2.Bowling_Type == strBowlingType].reset_index()
        dfBlue3 = dfBlue3[dfBlue3.Bowling_Type == strBowlingType].reset_index()
        dfMaroon = dfMaroon[dfMaroon.Bowling_Type == strBowlingType].reset_index()
        dfRed = dfRed[dfRed.Bowling_Type == strBowlingType].reset_index()
    
    batterColors = ['#FFA500', '#ed9640', '#0000FF', '#4f86e0', '#3a4f70', '#800000', '#FF0000']
    
    # Call to display batsmen analysis data    
    showAnalysis(dfOrange1,batterColors[0])
    showAnalysis(dfOrange2,batterColors[1])
    showAnalysis(dfBlue1,batterColors[2])
    showAnalysis(dfBlue2,batterColors[3])
    showAnalysis(dfBlue3,batterColors[4])
    showAnalysis(dfMaroon,batterColors[5])
    showAnalysis(dfRed,batterColors[6])
    
    plotGraphs()
```

```python
# Set dataframes for all batsmen
```

```python
def showAnalysis(dfBatsman, color):
    
    batterName = dfBatsman.Batsman_Name[0]
    ballsPlayed = len(dfBatsman)
    ballMiddledPer = round(100*dfBatsman.Ball_Middled.sum()/ballsPlayed, 2)
    ballMissedPer = round(100*dfBatsman.Ball_Missed.sum()/ballsPlayed, 2)
    ballEdgedPer = round(100*dfBatsman.Ball_Edged.sum()/ballsPlayed, 2)
    ballLeftPer = round(100*dfBatsman.Ball_Left.sum()/ballsPlayed, 2)
    
    aggressiveShotsPer = round(100*dfBatsman.Shot_Aggressive.sum()/ballsPlayed, 2)
    attackingShotsPer = round(100*dfBatsman.Shot_Attacking.sum()/ballsPlayed, 2)
    defensiveShotsPer = round(100*dfBatsman.Shot_Defensive.sum()/ballsPlayed, 2)
    
    properTimedShotsPer = round(100*dfBatsman.Shot_Timing.sum()/ballsPlayed, 2)
    properHeadPosPer = round(100*dfBatsman.Batsman_Head_Position.sum()/ballsPlayed, 2)
    properFootWorkPer = round(100*dfBatsman.Batsman_Footwork.sum()/ballsPlayed, 2)
    
    dfPercentages.loc[batterName] = [batterName, ballsPlayed, ballMiddledPer, ballMissedPer, ballEdgedPer, 
                                     ballLeftPer, aggressiveShotsPer, attackingShotsPer, defensiveShotsPer, 
                                     properTimedShotsPer, properHeadPosPer, properFootWorkPer, color]
    
    # Seperater    
    print('-------------------------------------------------------------')
    printmd('**BATTING ANALYSIS OF '+batterName+'**', color)
    
    printmd('**Total Balls Played : '+str(ballsPlayed)+'**', color)
    
    printmd('**MIDDLED** : **'+str(ballMiddledPer)+'** %', color)
    printmd('**MISSED** : **'+str(ballMissedPer)+'** %', color)
    printmd('**EDGED** : **'+str(ballEdgedPer)+'** %', color)
    printmd('**LEFT** : **'+str(ballLeftPer)+'** %', color)
    
    printmd('**AGGRESSIVE SHOTS** : **'+str(aggressiveShotsPer)+'** %', color)
    printmd('**ATTACKING SHOTS** : **'+str(attackingShotsPer)+'** %', color)
    printmd('**DEFENSIVE SHOTS** : **'+str(defensiveShotsPer)+'** %', color)
    
    printmd('**PROPER TIMED SHOTS** : **'+str(properTimedShotsPer)+'** %', color)
    printmd('**PROPER HEAD POSITION** : **'+str(properHeadPosPer)+'** %', color)
    printmd('**PROPER FOOTWORK** : **'+str(properFootWorkPer)+'** %', color)
    
```
