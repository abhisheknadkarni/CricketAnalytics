```python
import moviepy.editor as mp
import numpy as np
import librosa.display
import matplotlib.pyplot as plt
import random
```

```python
def random_color():
    return "#{:06x}".format(random.randint(0, 0xFFFFFF))
```

```python
# Display single audio channel
def display_audio_data(video_path):
    # Load video clip
    video_clip = mp.VideoFileClip(video_path)

    # Extract audio as numpy array
    audio_array = video_clip.audio.to_soundarray(fps=44100)

    # Plot the audio waveform
    plt.figure(figsize=(12, 4))
    librosa.display.waveshow(np.mean(audio_array, axis=1), sr=44100)
    plt.title('Audio Waveform')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()
```

```python
# Display multiple channel audio
def display_multi_channel_audio(video_path):
    # Load video clip
    video_clip = mp.VideoFileClip(video_path)

    # Extract audio
    audio = video_clip.audio

    # Convert audio to array
    audio_array = np.array(audio.to_soundarray(fps=44100))

    # Number of audio channels
    num_channels = audio_array.shape[1]

    # Plot each channel separately
    plt.figure(figsize=(12, 4 * num_channels))

    for i in range(num_channels):
        plt.subplot(num_channels, 1, i + 1)
        plt.plot(audio_array[:, i], color=random_color())
        plt.title(f'Audio Channel {i + 1}')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

    plt.tight_layout()
    plt.show()
```

```python
def display_two_channel_audio(video_path):
    # Load video clip
    video_clip = mp.VideoFileClip(video_path)

    # Extract audio as numpy array
    audio_array = np.array(video_clip.audio.to_soundarray(fps=44100))

    # Plot both audio channels in different colors
    plt.figure(figsize=(12, 4))
    plt.plot(audio_array[:, 0], label='Channel 1', color='blue')
    plt.plot(audio_array[:, 1], label='Channel 2', color='red')
    
    plt.title('Audio Waveform (Two Channels)')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()
```

```python
video_path = '../Media/Shot1.mp4'
display_audio_data(video_path)
```

```python
display_multi_channel_audio(video_path)
```

```python
display_two_channel_audio(video_path)
```
